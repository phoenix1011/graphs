```python
import csv
import numpy
from matplotlib import pyplot
from math import *
from collections import OrderedDict
from matplotlib.transforms import blended_transform_factory

%matplotlib inline
```

```python
x = numpy.linspace(190,1100,911)

y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []
y7 = []
y8 = []

t1 = []
t2 = []
t3 = []
t4 = []
t5 = []
t6 = []
t7 = []
```

```python
with open('CuO_19_Oct.csv','r') as infile:
    next(infile) # skip headings
    reader=csv.reader(infile,delimiter=',')
    for c1,c2,c3,c4,c5 in reader:
        y1.append(abs(float(c1)))
        y2.append(abs(float(c2)))
        y3.append(abs(float(c3)))
        y4.append(abs(float(c4)))
        y5.append(abs(float(c5)))
#         y6.append(float(c6))
        
        t1.append(-numpy.log(abs(float(c1))/100))
        t2.append(-numpy.log(abs(float(c2))/100))
        t3.append(-numpy.log(abs(float(c3))/100))
        t4.append(-numpy.log(abs(float(c4))/100))
        t5.append(-numpy.log(abs(float(c5)/100)))
#         t6.append(-numpy.log(float(c6)/100))


Y1 = numpy.array(y1)
Y2 = numpy.array(y2)
Y3 = numpy.array(y3)
Y4 = numpy.array(y4)
Y5 = numpy.array(y5)
# Y6 = numpy.array(y6)


T1 = numpy.array(t1)
T2 = numpy.array(t2)
T3 = numpy.array(t3)
T4 = numpy.array(t4)
T5 = numpy.array(t5)
# T6 = numpy.array(t6)
```

```python
linestyles = OrderedDict(
    [('solid',               (0, ())),
     ('loosely dotted',      (0, (1, 10))),
     ('dotted',              (0, (1, 5))),
     ('densely dotted',      (0, (1, 1))),

     ('loosely dashed',      (0, (5, 10))),
     ('dashed',              (0, (5, 5))),
     ('densely dashed',      (0, (5, 1))),

     ('loosely dashdotted',  (0, (3, 10, 1, 10))),
     ('dashdotted',          (0, (3, 5, 1, 5))),
     ('densely dashdotted',  (0, (3, 1, 1, 1))),

     ('loosely dashdotdotted', (0, (3, 10, 1, 10, 1, 10))),
     ('dashdotdotted',         (0, (3, 5, 1, 5, 1, 5))),
     ('densely dashdotdotted', (0, (3, 1, 1, 1, 1, 1)))])


pyplot.figure(figsize=(12,10))
pyplot.xlim(180, 1110)
pyplot.ylim(0, 100)

pyplot.tick_params(axis='both', labelsize=16)
pyplot.xlabel('Wavelength  $(nm)$', fontsize=16)
pyplot.ylabel('Transmittance   $(\%)$', fontsize=16)

pyplot.plot(x, Y1, linewidth=1.3, color = 'black', linestyle=(0, (7, 14)), marker='o', markevery=120, label="0.00005   vol.%")
pyplot.plot(x, Y2, linewidth=1.3, color = 'black', linestyle=(0, (1, 5)), marker='s',markevery=120, label="0.0001     vol.%")
pyplot.plot(x, Y3, linewidth=1.3, color = 'black', linestyle=(0, (7, 20, 2, 20, 2, 20)), marker='^',markevery=120, label="0.0005     vol.%")
pyplot.plot(x, Y4, linewidth=1.3, color = 'black', linestyle=(0, (2, 15)), marker='x',markevery=120, label="0.001       vol.%")
pyplot.plot(x, Y5, linewidth=1.3, color = 'black', linestyle=(0, (8, 5, 1, 5, 1, 5)), marker='v',markevery=120, label="0.005       vol.%")
# pyplot.plot(x, Y6, linewidth=1.3, color = 'black', linestyle=(0, (5, 25)), marker='^',markevery=120, label="0.0254   vol.%")

pyplot.legend(loc=4, prop={'size': 16}, frameon=False)

# pyplot.title('CuO - Water: Transmittance', fontsize=16)
pyplot.tight_layout()

pyplot.savefig('CuO_in_Water_19_Oct.png')
```
