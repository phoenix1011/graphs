```python
import csv
import numpy
from matplotlib import pyplot
from math import *
%matplotlib inline
```

```python
x = numpy.linspace(190,1100,911)

y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []
y7 = []
y8 = []
t1 = []
```

```python
with open('nan.csv','r') as infile:
    next(infile) # skip headings
    reader=csv.reader(infile,delimiter=',')
    for c1,c2,c3 in reader:
        y1.append(float(c1))
        y2.append(float(c2))
        y3.append(float(c3))

Y1 = numpy.array(y1)
Y2 = numpy.array(y2)
Y3 = numpy.array(y3)

T2 = numpy.ones(911)
print (T2)
T2 = -numpy.log(Y2/100)

print (T2)
```

```python
# pyplot.rc('font', family='serif', serif='Times')
# pyplot.rc('text', usetex=True)
# pyplot.rc('xtick', labelsize=8)
# pyplot.rc('ytick', labelsize=8)
# pyplot.rc('axes', labelsize=8)

pyplot.figure(figsize=(12,10))
# pyplot.ylim(-10,100)
# pyplot.xlim(190,1100)

pyplot.xlim(180, 1110)
pyplot.ylim(-3, Y1.max()*1.1)



pyplot.tick_params(axis='both', labelsize=18)
pyplot.xlabel('Wavelength (nm)', fontsize=18)
pyplot.ylabel('Transmittance (%)', fontsize=18)
pyplot.plot(x, Y1, color='red', linewidth=1.5, linestyle='-', label="Label 1")
# pyplot.plot(x, y2, color='blue', linewidth=1.5, linestyle='--')
pyplot.plot(x, y3, color='#aa0000', linewidth=2.5, linestyle=':', label="Label 2")
pyplot.plot(x, y2, color='black', linewidth=2, linestyle='-.', label="Label 3")

pyplot.legend(loc=4, prop={'size': 18}, frameon=False)

# pyplot.plot(x, y1, 'k-', x, y2, 'g', x, y3, 'r-.');
# pyplot.legend(['0.00025337 v%', '0.00053840 v%', '0.00101346 v%', '0.00498812 v%'], fontsize=16)
pyplot.title('CuO/Water', fontsize=18)
# pyplot.grid(color='#F5F5F5')

# pyplot.grid(color='#D2D2D2', linestyle='--', linewidth=1.4)

# pyplot.yaxis.tick_right()

pyplot.savefig('CuO_in_Water.png')
```
