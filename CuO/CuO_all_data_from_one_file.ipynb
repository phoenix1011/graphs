```python
import csv
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
x = numpy.linspace(190,1100,911)

y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []
y7 = []
y8 = []
```

```python
with open('nan.csv','r') as infile:
    next(infile) # skip headings
    reader=csv.reader(infile,delimiter=',')
    for c1,c2,c3 in reader:
        y1.append(float(c1))
        y2.append(float(c2))
        y3.append(float(c3))
```

```python
pyplot.figure(figsize=(10,10))
pyplot.ylim(0,100)
pyplot.tick_params(axis='both', labelsize=16)
pyplot.xlabel('wavelength (nm)', fontsize=16)
pyplot.ylabel('Transmittance (%)', fontsize=16)
pyplot.plot(x, y1, 'b--', x, y2, 'g', x, y3, 'r-.');
# pyplot.legend(['0.00025337 v%', '0.00053840 v%', '0.00101346 v%', '0.00498812 v%'], fontsize=16)
pyplot.title('CuO/Water', fontsize=16)
```
