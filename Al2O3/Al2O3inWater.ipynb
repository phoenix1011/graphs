**Transmission sprectra of Al2O3/water**

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
x = numpy.linspace(190,1100,911)

y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
```

Volume fraction = **0.0000875**
Color = **Blue**

```python
file_name = "Al2O3_0.7.txt"
with open(file_name, 'r') as infile:
    for line in infile:
        y1.append(float(line))
```

Volume fraction = **0.000125**
Color = **Green**

```python
file_name = "Al2O3_0.5.txt"
with open(file_name, 'r') as infile:
    for line in infile:
        y2.append(float(line))
```

Volume fraction = **0.000650**

```python
file_name = "Al2O3_1.3.txt"
with open(file_name, 'r') as infile:
    for line in infile:
        y3.append(float(line))
```

Volume fraction = **0.00125**

```python
file_name = "Al2O3_2.5.txt"
with open(file_name, 'r') as infile:
    for line in infile:
        y4.append(float(line))
```

```python
file_name = "Al2O3_20.txt"
with open(file_name, 'r') as infile:
    for line in infile:
        y5.append(float(line))
```

```python
pyplot.figure(figsize=(15,15))
pyplot.ylim(0,100)
pyplot.tick_params(axis='both', labelsize=16)
pyplot.xlabel('wavelength (nm)', fontsize=18)
pyplot.ylabel('Transmittance (%)', fontsize=18)
pyplot.plot(x, y1, 'b-', x, y2, 'g-', x, y3, 'r-', x, y5, 'c-', x, y4, 'm-');
pyplot.legend(['0.0000875 v%', '0.000125 v%', '0.000650 v%', '0.00125 v%', '0.005 v%'], fontsize=16)
pyplot.title('Al2O3/Water', fontsize=16)
```
