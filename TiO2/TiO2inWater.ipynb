```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
x = numpy.linspace(190,1100,911)

y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
```

Volume fraction = **0.00010206**

```python
file_name = "TiO2_1.txt"
with open(file_name, 'r') as infile:
    for line in infile:
        y1.append(float(line))
```

```python
file_name = "TiO2_2.1.txt"
with open(file_name, 'r') as infile:
    for line in infile:
        y2.append(float(line))
```

```python
pyplot.figure(figsize=(10,10))
pyplot.ylim(0,100)
pyplot.tick_params(axis='both', labelsize=16)
pyplot.xlabel('wavelength (nm)', fontsize=16)
pyplot.ylabel('Transmittance (%)', fontsize=16)
pyplot.plot(x, y1, 'b-', x, y2, 'g-', x, y3, 'r-', x, y4, 'c-');
pyplot.legend(['0.00010206 v%', '0.00049296 v%', '0.00103286 v%', '0.00502347 v%'], fontsize=16)
pyplot.title('TiO2/Water', fontsize=16)
```
